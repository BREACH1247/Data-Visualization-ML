```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=90)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

sns.pairplot(tips)

```python
sns.pairplot(tips,hue='sex')
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,color='black',hue='smoker')
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
tc  = tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```

```python
fp = flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(fp,linecolor = 'white',linewidth = '1')
```

```python
sns.clustermap(fp,cmap='coolwarm',linecolor = 'white',linewidth = '1',standard_scale=1)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```
